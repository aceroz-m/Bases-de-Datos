# **Grafico Altura vs Tiempo**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar datos
df = pd.read_csv("telemetry.csv")

plt.figure(figsize=(10,5))
plt.plot(df['time'], df['altitude'], linewidth=2)
plt.title("Altura vs Tiempo")
plt.xlabel("Tiempo (s)")
plt.ylabel("Altura (m)")

# Punto de apogeo
apogeo_idx = df['altitude'].idxmax()
plt.scatter(df['time'][apogeo_idx], df['altitude'][apogeo_idx], color='red')
plt.text(df['time'][apogeo_idx], df['altitude'][apogeo_idx],
         " Apogeo", color='red')

plt.grid(True)
plt.show()
```

# **Velocidad Vertical vs Tiempo**

```python
import numpy as np

df['vel'] = np.gradient(df['altitude'], df['time'])

plt.figure(figsize=(10,5))
plt.plot(df['time'], df['vel'], linewidth=2)
plt.title("Velocidad Vertical vs Tiempo")
plt.xlabel("Tiempo (s)")
plt.ylabel("Velocidad (m/s)")
plt.grid(True)

# Velocidad = 0 -> apogeo
zero_cross = df.iloc[(df['vel'] - 0).abs().argsort()[:1]]
plt.scatter(zero_cross['time'], zero_cross['vel'], color='red')
plt.text(zero_cross['time'].values[0], zero_cross['vel'].values[0],
         " Velocidad = 0\n(Apogeo)", color='red')

plt.show()
```

# **Aceleracion vs Tiempo**

```python
plt.figure(figsize=(10,5))
plt.plot(df['time'], df['acc'], linewidth=1.8)
plt.title("Aceleracin vs Tiempo")
plt.xlabel("Tiempo (s)")
plt.ylabel("Aceleracin (m/s)")
plt.grid(True)

plt.show()
```
